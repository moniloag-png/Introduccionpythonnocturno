# Comentarios

### ¿Qué son?

Texto contenido en ficheros de pyhton que es ignorado por el intérprete; es decir, no es ejecutado

### ¿Cuál es su utilidad?

* Se puede utilizar para documentar código y hacerlo más legible

* Preferiblemente trataremos de hacer código fácil de entender y que necesite pocos comentarios, en lugar de vernos forzados a recurrir a los comentarios para explicar el código

### Tipos de comentarios

**Comentarios de una línea**

* Texto precedido por "#"
* Se suele usar para documentar expresiones sencillas


```python
# Esto es una instrucción print
print("Hello world") # Esto es una instrucción print
```

**Comentarios de varia líneas**

* Texto encapsulado en triple comillas (que pueden ser simples como dobles)
* Se suele usar para documentar bloques de códigos más significativos

```python
# creo una función de multiplicación
def producto(x,y):
    """
    Esta función recibe dos números como parámetros y devuelve como resultado el producto de los mismos
    """
    #creo una variable donde guardo el producto de los números recibidos
    c=x*y
    return c    
```

```python
# Llamo a la función
producto(3,7)
```

# Literales, variables y tipos de datos básicos

De forma muy genérica, al ejecutarse un programa en Python, simplemente se realizan *operaciones* sobre *objetos*.

Estos dos términos son fundamentales.

* *Objetos*: Cualquier tipo de datos que sean (numéricos, caracteres o datos mas complejos)
  
* *Operaciones*: Como manipulamos estos datos, +, -, *, /

Ejemplo:
4*3

```python
4.3/3
```

## Literales

* Pyhton tiene una serie de tipos de datos que están integrados en el propio lenguaje.
* Los literales son expresiones que generan objetos de estos tipos.
* Estos objetos, según su tipo pueden ser:
  * Simples o compuestos
  * Mutables o inmutables

### Literales Simples
- Enteros
- Decimales
- Booleanos (V y F)

```python
print(4) # número entero
print(4.3) # número en coma flotante
print("Hello world!") #variable de tipo string
print(True) # Objeto de tipo booleano
```
