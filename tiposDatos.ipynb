### 2.2 Variables

- Referencias a objetos.
- Las variables y los objetos se almacenan en diferentes zonas de memoria.
- Las variables siempre referencian a objetos y nunca a otras variables.
- Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.
- Sentencia de asignación:
  
<nombre_variable> "=" <objeto>

```python
## Asignación de variables
a = 5
print(a)
```

### Operadores aritméticos

| Operador | Descripción |
| :--- | :--- |
| `a + b` | Suma |
| `a - b` | Resta |
| `a / b` | División |
| `a // b` | División Entera |
| `a % b` | Módulo / Resto |
| `a * b` | Multiplicación |
| `a ** b` | Exponenciación |

```python
x = 3
y = 2
print('x + y = ', x + y)
print('x - y = ', x - y)
print('x * y = ', x * y)
print('x / y = ', x / y)
print('x // y = ', x // y)
print('x % y = ', x % y)
print('x ** y = ', x ** y)
```

### Operadores de comparación

| Operador | Descripción |
| :--- | :--- |
| `a > b` | Mayor que |
| `a < b` | Menor que |
| `a == b` | Igualdad |
| `a != b` | Desigualdad (Diferente de) |
| `a >= b` | Mayor o Igual que |
| `a <= b` | Menor o Igual que |

```python
x = 10
y = 12
print('x > y es ', x > y)
print('x < y es ', x < y)
print('x == y es ', x == y)
print('x != y es ', x != y)
print('x >= y es ', x >= y)
print('x <= y es ', x <= y)
```

### Operadores Lógicos

| Operador | Descripción |
| :--- | :--- |
| `a and b` | True, si ambos son True |
| `a or b` | True, si alguno de los dos es True |
| `a ^ b` | XOR - True, si solo uno de los dos es True |
| `not a` | Negación (Invierte el valor) |

Enlace a [Tablas de Verdad](https://en.wikipedia.org/wiki/Truth_table)

```python
x = True
y = False
print('x and y es :', x and y)
print('x or y es :', x or y)
print('x xor y es :', x ^ y)
print('not x es :', not x)
```

* Son mutables y las asignaciones a un objeto alteran el primero
* Pasar listas a funciones puede suponer un riesgo
* Clonar o copiar listas

```python
lista = [2, 4, 16, 32]
ref = lista
print(id(lista))
print(id(ref))
ref[2] = 64
print(ref)
print(lista)
```

```python
lista = [2000, 4, 16, 32]
copia = lista[:]
copia = lista.copy()
copia[2] = 64
print(lista)
print(copia)
print(id(lista))
print(id(copia))
print(id(lista[2]))
print(id(copia[2]))
```

```python
# #listas anidadas se copian por referencia
lista = [2, 4, 16, 32, [34, 10, [5,5]]]
copia = lista.copy()
copia[3] = 20
copia[4][0] = 28
print(copia)
print(lista)
```

```python
lista2 = [10, 20]
lista = [0, 1, lista2]
copia = lista[:] # Se crea una copia superficial

print(f"Copia: {copia}") 
# Resultado: [0, 1, [10, 20]]

print(f"ID lista: {id(lista)}")
print(f"ID copia: {id(copia)}")
# Los IDs son DIFERENTES. Son dos objetos distintos en el primer nivel.

print(f"ID lista[2]: {id(lista[2])}")
print(f"ID copia[2]: {id(copia[2])}")
# ¡Los IDs son IGUALES! Ambos apuntan a la misma lista2 interna.

copia[2][0] = 40 # Modificamos la lista interna a través de la copia

print(f"Copia final: {copia}")
# Resultado: [0, 1, [40, 20]]

print(f"Lista original final: {lista}")
# Resultado: [0, 1, [40, 20]] <--- ¡TAMBIÉN CAMBIÓ!
```

```python
import copy

# 1. Definimos una lista con múltiples niveles de anidamiento
lista = [2, 4, 16, 32, [34, 10, [5, 5]]]

# 2. Realizamos una copia profunda (crea objetos nuevos para cada nivel)
copia = copy.deepcopy(lista)

# 3. Modificamos el primer nivel de la copia
copia[0] = 454

# 4. Modificamos el tercer nivel de profundidad (la lista dentro de la lista)
# Accedemos al índice 4, luego al índice 2, y cambiamos el valor en el índice 0
copia[4][2][0] = 64

# 5. Imprimimos resultados para comparar
print("Original:", lista)
print("Copia:   ", copia)

# 6. Comprobamos las direcciones de memoria (IDs)
print(f"IDs principales: {id(lista)} - {id(copia)}")
print(f"IDs nivel anidado: {id(lista[4])} - {id(copia[4])}")
```

## 3.3 Sets (Conjuntos)

Al igual que las listas:

* Colección de elementos.
* Tipos arbitrarios.
* Mutables.
* No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.

A diferencia de las listas:

* No puede tener duplicados.
* Se definen por medio de llaves.
* Los elementos no van ordenados por posición. No hay orden establecido.
* Solo pueden contener objetos inmutables.
* No soportan anidamiento.

### Operaciones con conjuntos

• Creación de conjuntos.

```python
set1 = {0, 1, 1, 2, 3, 4, 4}
print(set1)
21
set2 = {'user1', 12, 2}
print(set2)
set3 = set(range(7))
print(set3)
set4 = set([0, 1, 2, 3, 4, 0, 1])
print(set4)
```

```python
#Observa la diferencia entre listas y conjuntos
s = 'aabbc'
print(list(s))
print(set(s))
```

• Acceso por índice genera error.

```python
set1 = {0, 1, 2}
lista1 = list(set1) #Convertimos el set a lista
print(lista1[0])
```

• Unión, intersección y diferencia.

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4, 42])
# union
print(set1 | set2)
# intersección
print(set1 & set2)
# diferencia
print(set1 - set2)
print(set2 - set1)
```

```python
a = [1,2]
b = [2,3]
print(set(a) & set(b))
```

```python
#Además de los operadores, que operan únicamente con Sets, también se pueden usar métodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 1, 2}
lista = [1, 3, 3]
print(conjunto.union(lista))
print(conjunto.intersection(lista))
print(conjunto.difference(lista))
```

• Comparación de conjuntos.

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4])
print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
```

• Pertenencia.

```python
words = {'calm', 'balm'}
print('calm' in words)
```

• Anidamiento.

```python
# Los conjuntos no soportan anidamiento, pero como permite elementos inmutables, se pueden "anidar" tuplas.
nested_set = {1, (1, 1, 1), 2, 3, (1,1,1)}
print(nested_set)
```

• Modificación de conjuntos.

```python
# A través de operador de asignación
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1 |= set2 # set1 = set1 | set2
# set1 &= set2
set1 -= set2
print(set1)
```

```python
# A través de método 'update'.
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1.update(set2)
# set1.intersection_update(set2)
set1.difference_update(set2)
print(set1)
```

```python
# A través de métodos 'add' y 'remove'.
set1 = {'a', 'b', 'c'}
set1.add('d')
set1.remove('a')
print(set1)
```

### 3.4 Tuplas

Al igual que las listas:

* Colección de elementos.
* Tipos arbitrarios.
* Puede contener duplicados.
* No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.

```python
print("Ejercicio 7")

# Lista original
lista_inicial = [1, 2, 3, 4, 5, 6, 7, 8]

# Lista deseada [8, 6, 4, 2]
# Explicación del slice [::-2]:
# - El primer vacío significa: empezar desde el principio (al ser paso negativo, es el final).
# - El segundo vacío significa: terminar al final.
# - El -2 significa: avanzar hacia atrás de dos en dos.
lista_final = lista_inicial[::-2]

# Resultado:
print(lista_final)
```

```python
print("Ejercicio 8")

# Definir la tupla y el índice a eliminar
mi_tupla = (10, 20, 30, 40, 50)
i = 2  # Queremos eliminar el 30 (posición 2)

# Crear una nueva tupla uniendo:
# - Desde el inicio hasta antes de i: mi_tupla[:i]
# - Desde después de i hasta el final: mi_tupla[i+1:]
tupla_resultante = mi_tupla[:i] + mi_tupla[i+1:]

print("Tupla original:", mi_tupla)
print(f"Tupla tras eliminar índice {i}:", tupla_resultante)
```

```python
print("Ejercicio 9")

L = [7, 3, 1, 9, 5, 8, 2]

# 1. Ordenar lista de menor a mayor
L.sort()

# 2. Calcular la longitud y el índice central
n = len(L)
indice_central = n // 2  # El operador // hace división entera

# 3. Lógica para la mediana
if n % 2 != 0:
    # Si el número de elementos es impar, es el valor del medio
    mediana = L[indice_central]
else:
    # Si es par, según el ejercicio tomamos una (usualmente el promedio de las dos centrales)
    # Aquí tomaremos la que está justo en la mitad superior por simplicidad
    mediana = L[indice_central]

print("Lista ordenada:", L)
print("La mediana es:", mediana)
```
