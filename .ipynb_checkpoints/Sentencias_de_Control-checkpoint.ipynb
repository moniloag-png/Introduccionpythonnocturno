# 1. Estructuras de control
## 1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
# # Asigno el valor uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')      
else:
    print('Valor negativo')
```

## 2. Swith

```python
a = -6
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```python
a = 'l' #week day
match a:
    case 'L':
        print("Lunes")
    case 'M':
       print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

Crear un programa en Python donde el usuario ingrese el tipo de operación aritmética a realizar, ejemplo, suma, resta, multiplicación y división. Además el usuario debe ingresar dos números enteros y realizar la operación que ingresó el usuario. Para el programa solo se debe utilizar el condicional match

```python
# Solicitar el tipo de operación
operacion = input("Ingrese la operación (suma, resta, multiplicacion, division): ").lower()

# Solicitar los dos números enteros
num1 = int(input("Ingrese el primer número entero: "))
num2 = int(input("Ingrese el segundo número entero: "))

# Realizar la operación usando match
match operacion:
    case "suma":
        resultado = num1 + num2
        print(f"Resultado: {num1} + {num2} = {resultado}")
    case "resta":
        resultado = num1 - num2
        print(f"Resultado: {num1} - {num2} = {resultado}")
    case "multiplicacion":
        resultado = num1 * num2
        print(f"Resultado: {num1} * {num2} = {resultado}")
    case "division":
        if num2 != 0:
            resultado = num1 / num2
            print(f"Resultado: {num1} / {num2} = {resultado}")
        else:
            print("Error: No se puede dividir entre cero.")
    case _:
        print("Operación no válida.")
```

## Expresión ternaria
* Estructura if-else condensado en una línea.
* Se recomienda usar solo en casos sencillos.

```python
# # # Ejemplo: cálculo del valor absoluto de un número.
x = -3
resultado = x if x >= 0 else -x
print(resultado)
```

```python
val = -3
if val >=0:
    resultado = val
else:
    resultado = -val

print(resultado)
```
