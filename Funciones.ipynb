## 1.1 ¿Qué son?
* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera

* En cada invocación de la función, los parámetros de entrada pueden variar.

### Ejercicio:
Sabiendo que la fórmula de la ecuación general de segundo grado es X12=-b+-raiz(b2-4*a*c)/2*a. Crear una función donde el usuario mediante el teclado ingrese los valores de a, b y c. Calcular las raices de la ecuación.

````python
##### import math

def calcular_raices():
    # Ingreso de datos
    a = float(input("Ingrese el valor de a: "))
    b = float(input("Ingrese el valor de b: "))
    c = float(input("Ingrese el valor de c: "))

    # Validar que sea una ecuación de segundo grado
    if a == 0:
        print("Error: el valor de 'a' no puede ser 0.")
        return

    # Calcular el discriminante
    discriminante = b**2 - 4*a*c

    # Analizar el discriminante
    if discriminante > 0:
        x1 = (-b + math.sqrt(discriminante)) / (2*a)
        x2 = (-b - math.sqrt(discriminante)) / (2*a)
        print("La ecuación tiene dos raíces reales diferentes:")
        print("x1 =", x1)
        print("x2 =", x2)

    elif discriminante == 0:
        x = -b / (2*a)
        print("La ecuación tiene una raíz real doble:")
        print("x =", x)

    else:
        print("La ecuación no tiene raíces reales (discriminante negativo).")

# Llamada a la función
calcular_raices()
````

## 1.8 Funciones Lambda
* Es posible definir funciones anónimas (sin nombre).
* Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia *def*
no puede (por ejemplo, dentro de una lista o como parámetro de una función).
* Útiles cuando se quiere pasar una función como argumento a otra.
#### Formato general:
lambda <lista de argumentos> : <valor a retornar>
* Importante: <valor a retornar> no es un conjunto de instrucciones. Es simplemente una
expresión return que omite esta palabra.
* Las funciones definidas con def son más generales:
    * Cualquier cosa que implementes en un lambda, lo puedes implementar como una función
convencional con def, pero no viceversa.

````python
# Ejemplo: función suma.
def suma(x, y, z):
    return x + y + z
print(suma(1, 2, 3))

# Ejemplo: función suma como expresión lambda.

suma2 = lambda x, y, z : x + y + z
print(suma2(1, 2, 3))
````

````python
#La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

cities.sort()
print(cities)

cities.sort(key = lambda x : len(x))
print(cities)

cities.sort(key = len)
print(cities)

def longitud(x):
    return len(x)
cities.sort(key = longitud)
print(cities)
````

````python
# Mismo ejemplo sin lambda
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
def num_caracteres(x):
    return len(x)
cities.sort(key = num_caracteres)
print(cities)
````

* Nos podemos guardar una referencia para utilización repetida.

````python
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
f = lambda x : len(x)
for s in cities:
    print(f(s))
````

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* Ideal para escoger entre varias opciones

````python
def switch_dict(operator, x, y):
# crea un diccionario de opciones y selecciona 'operator'
    return {
    'add': lambda: x + y,
    'sub': lambda: x - y,
    'mul': lambda: x * y,
    'div': lambda: x / y,
}.get(operator, lambda: None)() # retorna None si no encuentra 'operator'
print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
````

## 1.9 Scopes
* La localización de una asignación a una variable en el código determina desde donde puedes
acceder a esa variable.
* Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
* El alcance de la visibilidad de una variable determina su scope.
* El término scope hace referencia a un espacio de nombres (namespace). Por ejemplo, una
función establece su propio namespace.

````python
X = 10

def func():
    X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
    
    def func_int():
        X = 30
        print("El valor de X es:", X)
    func_int()
func()
````

#### Regla LEGB Dada una función F, tenemos los siguientes scopes:
* Local: variables locales a F.
* Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
* Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función.
Este scope no abarca más de un fichero.
* Built-ins: proporcionadas por el lenguaje.
La resolución de nombres ocurre de abajo a arriba.

La resolución de nombres ocurre de abajo a arriba.

````python
# Global (X y func)
X = 99

def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z

print(func(1))
````

````python
X = 1

def func():
    X = 2 # X es una variable diferente

func()
print(X)
````

* La sentencia global nos permite modificar una variable global desde dentro de una función.

````python
X = 1

def func():
    global X
    X = 2
func()
print(X)
````

````python
y, z = 1, 2
def todas_globales():
    global x
    x = y + z # Las tres variables son globales.
todas_globales()
print(x)
print(y)
print(z)
````

````python
y, z = 1, 2
def todas_globales():
    global x
x = y + z # Las tres variables son globales.
todas_globales()
print(x)
print(y)
print(z)
````

````python
X = 77
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
            print(X)
    f2()
f1()
````

* Closures
    * Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan
existiendo o no.

````python
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla
accion = f1()
accion()
````

* Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

````python
def f1():
    contador = 10
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2
accion = f1()
accion()
accion()
accion()
````

````python
print("Ejercicio 8:")

# Definición de la función lambda
tiene_duplicados = lambda lista: len(lista) != len(set(lista))

# --- Ejemplos de uso ---
lista_a = [1, 2, 3, 4, 5]
lista_b = [1, 2, 2, 3, 4]

print(f"Lista A tiene duplicados: {tiene_duplicados(lista_a)}") # False
print(f"Lista B tiene duplicados: {tiene_duplicados(lista_b)}") # True
````

````python
print("Ejercicio 9:")

def es_palindromo(texto):
    # 1. Limpieza (opcional pero recomendada): 
    # Convertimos a minúsculas para que 'Reconocer' sea válido
    texto = texto.lower()
    
    # 2. Comparamos el texto con su versión invertida [[::-1]]
    return texto == texto[::-1]

# Pruebas
print(es_palindromo("reconocer"))  # True
print(es_palindromo("python"))     # False
````
