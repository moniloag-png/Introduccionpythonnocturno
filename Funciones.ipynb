# 1 Funciones
## 1.1 ¿Qué son?
* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera
que se puedan ejecutar cuando el programador considere oportuno.
* Las funciones van identificadas por un nombre.
* Se debe usar este nombre para invocar (ejecutar) el código de la función.
* Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
* En cada invocación de la función, los parámetros de entrada pueden variar.
## 1.2 Ejemplo

```python
def sumar(x, y):
    suma=x+y
    return suma
print(sumar(4,5))
print(sumar(-2,0))
```

## 1.3 ¿Qué ventajas nos proporcionan?
* Eliminación de duplicación de código.
* Incremento de la reutilización de código.
* Manejo de complejidad: permiten descomponer grandes sistemas en piezas más pequeñas y,
por tanto, más manejables y comprensibles.
## 1.4 Programación de funciones en Python
Las funciones se definen por medio de la palabra clave *def*.

Formato general:
def name(arg1, arg2, ..., argN):
statements

Muy comúnmente:
def name(arg1, arg2, ..., argN):
...
return value
* Las funciones son instrucciones convencionales, que se ejecutan cuando el flujo de control
llega a ellas.
* La definición de la función debe ejecutarse antes de poder invocarse la función.

```python
# Primero, definición.
def crear_diccionario(keys, values):
    return dict(zip(keys, values))
```

```python
# # Después, invocación.
nombres = ['Alfred', 'James', 'Peter', 'Harvey']
edades = [87, 43, 19, 16]
usuarios = crear_diccionario(nombres, edades)
print(usuarios)
```

* Las funciones pueden estar anidadas en sentencias condicionales o bucles.

```python
a = -2 # fijarse siempre en la condición
if a > 0:
    def operacion(x, y):
        return x + y
else:
    def operacion(x, y):
        return x * y
print(operacion(3, 4))
```

### Ejercicio:
Sabiendo que la fórmula de la ecuación general de segundo grado es X12=-b+-raiz(b2-4*a*c)/2*a. Crear una función donde el usuario mediante el teclado ingrese los valores de a, b y c. Calcular las raices de la ecuación.

```python
import math

def calcular_ecuacion():
    # Pedir datos al usuario
    a = float(input("Valor de a: "))
    b = float(input("Valor de b: "))
    c = float(input("Valor de c: "))

    # Calcular el discriminante (lo que va dentro de la raíz)
    d = b**2 - 4*a*c

    # Calcular las dos soluciones
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)

    print(f"x1 es: {x1}")
    print(f"x2 es: {x2}")

# Ejecutar la función
calcular_ecuacion()
```

```python
##### import math

def calcular_raices():
    # Ingreso de datos
    a = float(input("Ingrese el valor de a: "))
    b = float(input("Ingrese el valor de b: "))
    c = float(input("Ingrese el valor de c: "))

    # Validar que sea una ecuación de segundo grado
    if a == 0:
        print("Error: el valor de 'a' no puede ser 0.")
        return

    # Calcular el discriminante
    discriminante = b**2 - 4*a*c

    # Analizar el discriminante
    if discriminante > 0:
        x1 = (-b + math.sqrt(discriminante)) / (2*a)
        x2 = (-b - math.sqrt(discriminante)) / (2*a)
        print("La ecuación tiene dos raíces reales diferentes:")
        print("x1 =", x1)
        print("x2 =", x2)

    elif discriminante == 0:
        x = -b / (2*a)
        print("La ecuación tiene una raíz real doble:")
        print("x =", x)

    else:
        print("La ecuación no tiene raíces reales (discriminante negativo).")

# Llamada a la función
calcular_raices()
```
