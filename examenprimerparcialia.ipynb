1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
# Programa para calcular el IMC

print("Datos del usuario")
peso = float(input("Ingresa tu peso en kg: "))
estatura = float(input("Ingresa tu estatura en metros: "))
imc = peso / (estatura **2)
print("Tu IMC es:", imc)
```

Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y2,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz


```python
import math

print("Coordenadas del primer punto:")
x1 = float(input("Ingresa x1: "))
y1 = float(input("Ingresa y1: "))

print("\nCoordenadas del segundo punto:")
x2 = float(input("Ingresa x2: "))
y2 = float(input("Ingresa y2: "))

print("Calcular la distancia usando la fórmula")
# math.sqrt() para la raíz y **2 para elevar al cuadrado
distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print(f"\nLa distancia entre los puntos es: {distancia:.4f}")
```
