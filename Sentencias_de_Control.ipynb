# 1. Estructuras de control
## 1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
# # Asigno el valor uno a la variable x
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]
print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')      
else:
    print('Valor negativo')
```

Además el usuario debe ingresar dos números enteros y realizar la operación que ingresó el usuario. Para el programa solo se debe utilizar el condicional match

```python
# Solicitar el tipo de operación
operacion = input("Ingrese la operación (suma, resta, multiplicacion, division): ").lower()

# Solicitar los dos números enteros
num1 = int(input("Ingrese el primer número entero: "))
num2 = int(input("Ingrese el segundo número entero: "))

# Realizar la operación usando match
match operacion:
    case "suma":
        resultado = num1 + num2
        print(f"Resultado: {num1} + {num2} = {resultado}")
    case "resta":
        resultado = num1 - num2
        print(f"Resultado: {num1} - {num2} = {resultado}")
    case "multiplicacion":
        resultado = num1 * num2
        print(f"Resultado: {num1} * {num2} = {resultado}")
    case "division":
        if num2 != 0:
            resultado = num1 / num2
            print(f"Resultado: {num1} / {num2} = {resultado}")
        else:
            print("Error: No se puede dividir entre cero.")
    case _:
        print("Operación no válida.")
```

### Valor booleano
* Todos los objetos en Python tiene inherentemente un valor booleano: *True o False*.
* Cualquier número distinto de 0 ó cualquier objeto no vacío tienen valor *True*.
* El número 0, objetos vacíos y el objeto especial None tienen valor *False*.

```python
a = -10 #True
b = None #False
if a:
    print("a es True")
if not b:
    print("b es False")
```

# 3 Iteración (bucles)
* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: while y for.

### Bucle ‘while’
* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

#### Formato general:
* while test:  # Mientras se cumple la condición
* statements   # Instrucciones a ejecutar

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

#### Ejercicio:

Desarrollar un progama en Python donde se pida al usuario un número entero positivo y muestre en pantalla todos los números impares desde el 1 hasta ese número separado por comas

```python
n = int(input("Introduce un número entero positivo: "))
i=1
# Guardamos el resultado en una lista
resultado = [] 
while i <= n:
    if i % 2 != 0:        # Verificamos si el número es impar (cuando el resto de la división es distinto a 0)
        resultado.append(str(i))
    i += 1                # Incrementamos el contador en cada vuelta
print(", ".join(resultado))
```

#### Ejercicio:

Crear un programa en Python donde el usuario ingrese un número por teclado, el programa le pregunte al usuario si desea continuar con el programa si o no. Si la respuesta es SI, el programa debe volver a ingresar el número. Si el usuario ingresa NO el programa termina.


```python
while True:
    try:
        numero = int(input("Ingresa un número entero: "))
        print(f"Has ingresado el número: {numero}")
    except ValueError:
        print("¡Error! Por favor, ingresa un número válido.")
        continue # Regresa al inicio del ciclo

    decision = input("¿Deseas continuar? (SI/NO): ").strip().upper()
    if decision == "NO":
        break
```

```python
# Empezamos con la variable en "SI" para que el ciclo entre la primera vez
respuesta = "SI"

while respuesta == "SI":
    # Pedimos el número
    numero = int(input("Ingresa un número: "))
    
    # Preguntamos si quiere seguir
    # Usamos .upper() para que acepte "si", "Si" o "SI"
    respuesta = input("¿Deseas continuar? (SI/NO): ").upper()

print("Programa terminado.")
```
